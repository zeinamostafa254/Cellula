```
#encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['booking status'] = le.fit_transform(df['booking status'])
df['market segment type'] = le.fit_transform(df['market segment type'])
df['type of meal'] = le.fit_transform(df['type of meal'])
df['room type'] = le.fit_transform(df['room type'])
```

```
df.drop(['Booking_ID', 'date of reservation', 'repeated', 'P-C', 'P-not-C'], axis = 1,inplace = True)
```

```
df
```

```
#check outliers
sns.boxplot(df, palette= 'rainbow', orient = 'h' )
```

```
#handling outliers
Q1 = df['lead time'].quantile(0.25)
Q3 = df['lead time'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['lead time'] >= lower_bound) & (df['lead time'] <= upper_bound)]
Q1 = df['average price '].quantile(0.25)
Q3 = df['average price '].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['average price '] >= lower_bound) & (df['average price '] <= upper_bound)]
```

```
sns.boxplot(df, palette= 'rainbow', orient = 'h' )
```

```
initial_rows = df.shape[0]

# Identify numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns

print(f"Initial number of rows: {initial_rows}")

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out outliers for the current column
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    print(f"After processing '{col}': {df.shape[0]} rows remaining")

final_rows = df.shape[0]
rows_removed = initial_rows - final_rows

print(f"\nFinal number of rows after removing outliers from all numerical columns: {final_rows}")
print(f"Total rows removed: {rows_removed}")

# Display the boxplot again to check for remaining outliers
sns.boxplot(df, palette= 'rainbow', orient = 'h' )
```

```
# Sum the data in each column
column_sums = df.sum()

# Display the sums
print("Sum of data in each column:")
print(column_sums)
```

```
df.drop(['car parking space', 'room type', 'number of children'], axis = 1,inplace = True)
```

```
#correlation
corr_matrix = df.corr(numeric_only=True).round(3)
```

```
#correlation heatmap
sns.heatmap(corr_matrix, annot = True , annot_kws = {"size":8},cmap ='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```
#create a figure paired with axis
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
fig.set_facecolor('lightgrey')

# Plotting distributions for selected columns
sns.histplot(df['market segment type'], ax=ax[0, 0], kde=True, color='blue')
ax[0, 0].set_title('Distribution of market segment type')

sns.histplot(df['number of week nights'], ax=ax[0, 1], kde=True, color='black')
ax[0, 1].set_title('Distribution of Number of Week Nights')

sns.histplot(df['average price '], ax=ax[1, 0], kde=True, color='darkgreen')
ax[1, 0].set_title('Distribution of Average Price')

sns.histplot(df['lead time'], ax=ax[1, 1], kde=True, color='orange')
ax[1, 1].set_title('Distribution of Lead Time')

plt.tight_layout()
plt.show()
```
